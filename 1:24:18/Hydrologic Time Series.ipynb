```python
import numpy as np # Computations
import pandas as pd # Data management
import matplotlib.pyplot as plt # Plotting

df = pd.read_csv('MF_Boise.csv', names=['Y', 'M', 'D', 'Q (cms)'], skiprows=1)
df[0:20]
```

```python
df['SerDates'] = pd.to_datetime(df.Y*10000+df.M*100+df.D, format='%Y%m%d')
df[0:20]
```

```python
df.plot(x='SerDates', y='Q (cms)', figsize=(14,10))
plt.show()
```

```python
Q = df['Q (cms)'].values
Qt = Q[1:-1]
Qtm1 = Q[0:-2]

plt.figure(figsize=(14,10))
plt.plot(Qtm1, Qt, 'o')
plt.title('How correlated is today\'s discharge with yesterday\'s?')
plt.xlabel('$Q_{t-1}$ m$^3$/s')
plt.ylabel('$Q_t$ m$^3$/s')
plt.show()
```
