## 1. Load the required libraries, load the data, double-check

```python
import numpy as np # Computations
import pandas as pd # Data management
import matplotlib.pyplot as plt # Plotting

# Load MF Boise discharge dataset
df = pd.read_pickle('MFBoiseDischarge_WY2001-2017.pkl')

# Print the first and last 10 records to make sure data is sufficient
qrows = np.concatenate([np.arange(0,10,1),np.arange(-11,-1,1)])
df.iloc[qrows,:]
```

## 2. How Correlated is Discharge Today with Discharge Yesterday?

```python
Q = df['Q (cms)'].values

# Use numpy array indexing to get discharge from October 2, 2000 to September 30, 2017. 
# The 'end' of a numpy array can be indexed using -1: as if the index in the negative direction
# wraps back around to the end of the record
Qt = Q[1:-1]

# Similarly get discharge from October 1, 2000 to September 29, 2017 by getting data from the 
# beginning of the record to the 'end' minus 1 using the -2 as the index. 
Qtm1 = Q[0:-2]

# Use Matplotlib to create a scatter plot of these two time series, create a title and label axes
plt.figure(figsize=(14,10))
plt.plot(Qtm1,Qt,'o')
plt.title('How correlated is today\'s discharge with yesterday\'s?',fontsize=16)
plt.xlabel('$Q_{t-1}$ m$^3$/s',fontsize=16)
plt.ylabel('$Q_t$ m$^3$/s',fontsize=16)
plt.show()
```

## 3. Get the Statistical Measure of This Correlation

```python
R = np.corrcoef(Qtm1,Qt)
R[1,0]
```

## 4. How Correlated is Discharge Today with Discharge 40 Days Ago?

```python
Qt   = Q[40:-1] 
Qtmk = Q[0:-(41)]

# Use Matplotlib to create a scatter plot of these two time series, create a title and label axes
plt.figure(figsize=(14,10))
plt.plot(Qtmk,Qt,'o')
plt.title('How correlated is discharge today with discharge 40 days ago',fontsize=16)
plt.xlabel('$Q_{t-40}$ m$^3$/s',fontsize=16)
plt.ylabel('$Q_t$ m$^3$/s',fontsize=16)
plt.show()
```

## 5. Get the statistical measure of this correlation (again)

```python
R = np.corrcoef(Qtmk, Qt)
R[1,0]
```

## 6. Get the Autocorrelation Coefficient for Lags from 0 to 40 days

```python
Nk = 40
K  = np.arange(0,Nk,1)
Rk = np.zeros((Nk,1))

for k in K:
    Qt   = Q[k:-1] 
    Qtmk = Q[0:-(1+k)]
    R = np.corrcoef(Qtmk,Qt)
    Rk[k] = R[1,0]
  

plt.figure(figsize=(14,10))
plt.plot(K,Rk,'o')
plt.xlim(0.0,np.amax(K.astype(float)))
plt.ylim(0.0,1.1)
plt.title('Autocorrelation function of MF Boise discharge',fontsize=16)
plt.xlabel('Lag distance [days]',fontsize=16)
plt.ylabel('Autocorrelation coefficient [-]',fontsize=16)
plt.show()
```

## 7. Pandas has a function for Autocorrelation Plots

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(df['Q (cms)'])
plt.show()
```

## 8. Compute the Autocorrelation Function a slightly different way

```python
Qp = Q - np.mean(Q)
autocovs2 = np.correlate(Qp, Qp, mode='full')
N = autocovs2.size
autocovs2_oneside = autocovs2[N//2:]
autocorrs2_oneside = autocovs2_oneside/autocovs2_oneside[0]

plt.figure(figsize=(14,10))
plt.plot(K,autocorrs2_oneside[0:Nk],'o')
plt.ylim(0.0,1.1)
plt.xlim(0.0,np.amax(K.astype(float)))
plt.title('Autocorrelation function of MF Boise discharge',fontsize=16)
plt.xlabel('Lag distance [days]',fontsize=16)
plt.ylabel('Autocorrelation coefficient [-]',fontsize=16)
plt.show()
```

## 9. Taking a look beyond 40 days to 7 years

```python
Nk2 = 365*7
K2 = np.arange(0,Nk2,1)

plt.figure(figsize=(14,10))
plt.plot(K2,autocorrs2_oneside[0:Nk2],'-')
plt.ylim(-1.0,1.0)
plt.xlim(0.0,np.amax(K2.astype(float)))
plt.title('Autocorrelation function of MF Boise discharge',fontsize=16)
plt.xlabel('Lag distance [days]',fontsize=16)
plt.ylabel('Autocorrelation coefficient [-]',fontsize=16)
plt.show()
```
