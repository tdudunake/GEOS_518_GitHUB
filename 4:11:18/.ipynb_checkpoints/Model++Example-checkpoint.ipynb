# Three Bucket Hydrology Model
## April 11, 2018

```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd 

df = pd.read_csv('2022_FortRenoOK_2010.csv', names=['Y','M','D','P','2','4','8','20','40'], skiprows=1) # import data
```

```python
Nt = len(df)
dt = 1.0

P = df['P'].values    
t = np.arange(1, Nt+1, 1)

print(P.shape)
print(t.shape)
```

```python
plt.figure(1)
plt.figure(figsize=(10,5)) 
plt.bar(t, P)
plt.ylabel('Precip [mm/d]')
plt.xlabel('time [d]')
plt.show()
```

```python
# BUCKET 1
k12 = 0.02 #Drainage coefficient in units of day^-1
k23 = 0.019
k3 = 0.018
W1_0 = 250.0  #Water storage in units of mm

# Initializing a data container for water storage at each time step
W1 = np.zeros(t.shape)
W2 = np.zeros(t.shape)
W3 = np.zeros(t.shape)

# Update initial condition
W1[0] = W1_0
W2[0] = W1_0
W3[0] = W1_0

# Initializing a data container for our discharge at each time step
Q12 = np.zeros(t.shape)
Q23 = np.zeros(t.shape)
Qb = np.zeros(t.shape)

# Update initial condition
Q12[0] = k12*W1[0]
Q23[0] = k23*W1[0]
Qb[0] = k3*W1[0]

# The main loop, bucket 1
for i in np.arange(1, Nt, 1):
    
    # Compute the value of the derivatives
    dW1dt = P[i-1] - k12*W1[i-1]
    dW2dt = k12*W1[i-1] - k23*W2[i-1]
    dW3dt = k23*W2[i-1] - k3*W3[i-1]
    
    # Compute the next value of W
    W1[i] = W1[i-1] + dW1dt*dt
    W2[i] = W2[i-1] + dW2dt*dt
    W3[i] = W3[i-1] + dW3dt*dt
    
    # Compute the next value of Q
    Q12[i] = k12*W1[i]
    Q23[i] = k23*W2[i]
    Qb[i] = k3*W3[i]
```

```python
# plt.figure(2)
plt.figure(2)
plt.figure(figsize=(12,10)) 
plt.subplot(311)
plt.bar(t, P)
plt.xlabel('Date')
plt.ylabel('Precip')
plt.subplot(312)
plt.plot(t, W1)
plt.xlabel('Date')
plt.ylabel('Storage')
plt.subplot(313)
plt.plot(t, Qb)
plt.xlabel('Date')
plt.ylabel('Discharge')
plt.show()
```
